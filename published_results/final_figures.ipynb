```python
import kaleido
import pandas as pan
import plotly
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio
```

```python
network_data = pan.read_csv('network_table_data.csv')

# subset network data in mortality 3 (low mortality for plants and pollinators) and attachment algorithm 3 (random)
network_data_m3 = network_data[network_data.death_case == 3]
network_data_m3_l1 = network_data_m3[network_data_m3.InvP_k_alg == 1]

# subset for each invader type
network_data_m3_l1_i1 = network_data_m3_l1[network_data_m3_l1.inv_type == 1]
network_data_m3_l1_i2 = network_data_m3_l1[network_data_m3_l1.inv_type == 2]
network_data_m3_l1_i3 = network_data_m3_l1[network_data_m3_l1.inv_type == 3]
network_data_m3_l1_i4 = network_data_m3_l1[network_data_m3_l1.inv_type == 4]
network_data_m3_l1_i5 = network_data_m3_l1[network_data_m3_l1.inv_type == 5]
network_data_m3_l1_i6 = network_data_m3_l1[network_data_m3_l1.inv_type == 6]
network_data_m3_l1_i7 = network_data_m3_l1[network_data_m3_l1.inv_type == 7]
network_data_m3_l1_i8 = network_data_m3_l1[network_data_m3_l1.inv_type == 8]
```

```python
# calculate proportion of invasions for each invader type
i1_count = round(len(network_data_m3_l1_i1[network_data_m3_l1_i1.InvP_inv == 1]) / len(network_data_m3_l1_i1), 2)
i2_count = round(len(network_data_m3_l1_i2[network_data_m3_l1_i2.InvP_inv == 1]) / len(network_data_m3_l1_i2), 2)
i3_count = round(len(network_data_m3_l1_i3[network_data_m3_l1_i3.InvP_inv == 1]) / len(network_data_m3_l1_i3), 2)
i4_count = round(len(network_data_m3_l1_i4[network_data_m3_l1_i4.InvP_inv == 1]) / len(network_data_m3_l1_i4), 2)
i5_count = round(len(network_data_m3_l1_i5[network_data_m3_l1_i5.InvP_inv == 1]) / len(network_data_m3_l1_i5), 2)
i6_count = round(len(network_data_m3_l1_i6[network_data_m3_l1_i6.InvP_inv == 1]) / len(network_data_m3_l1_i6), 2)
i7_count = round(len(network_data_m3_l1_i7[network_data_m3_l1_i7.InvP_inv == 1]) / len(network_data_m3_l1_i7), 2)
i8_count = round(len(network_data_m3_l1_i8[network_data_m3_l1_i8.InvP_inv == 1]) / len(network_data_m3_l1_i8), 2)

# plot these proportions in a bar graph
inv_type = ["Specialist", "Spec High R", "Spec High P", "Spec High R&P", "Generalist", "Gen High R", "Gen High P", "Gen High R&P"]
inv_count = [i1_count, i3_count, i2_count, i4_count, i5_count, i7_count, i6_count, i8_count]
fig = go.Figure(data = go.Bar(y = inv_count, x = inv_type, text = inv_count, textposition = 'outside'))
fig.update_layout(plot_bgcolor = 'rgba(0,0,0,0)', font = dict(size = 20, color = 'black'), width = 1000, height = 700, margin=dict(t = 100))
fig.update_traces(marker_color='black')
fig.update_xaxes(showline=True, linecolor='black')
fig.update_yaxes(showline=True, linecolor='black', title = "Proportion of Successful Invasions", range = [0,1])
fig.show()

fig.write_image('fig1a.png', width = 800, height = 650, scale = 10)
```

```python
# histrograms of total initial pollinator abundance

#for invader type 3
sum_init_A_inv3 = network_data_m3_l1_i3[network_data_m3_l1_i3.InvP_inv == 1]
sum_init_A_no_inv3 = network_data_m3_l1_i3[network_data_m3_l1_i3.InvP_inv == 0]
sum_init_A_inv3 = list(sum_init_A_inv3['init_A'])
sum_init_A_no_inv3 = list(sum_init_A_no_inv3['init_A'])

#for invader type 4
sum_init_A_inv4 = network_data_m3_l1_i4[network_data_m3_l1_i4.InvP_inv == 1]
sum_init_A_no_inv4 = network_data_m3_l1_i4[network_data_m3_l1_i4.InvP_inv == 0]
sum_init_A_inv4 = list(sum_init_A_inv4['init_A'])
sum_init_A_no_inv4 = list(sum_init_A_no_inv4['init_A'])

#for invader type 8
sum_init_A_inv8 = network_data_m3_l1_i8[network_data_m3_l1_i8.InvP_inv == 1]
sum_init_A_no_inv8 = network_data_m3_l1_i8[network_data_m3_l1_i8.InvP_inv == 0]
sum_init_A_inv8 = list(sum_init_A_inv8['init_A'])
sum_init_A_no_inv8 = list(sum_init_A_no_inv8['init_A'])

fig = make_subplots(rows=3, cols=1, shared_xaxes=True, shared_yaxes = True)                                                   

# add all histograms
fig.add_trace(go.Histogram(x = sum_init_A_inv3, nbinsx = 30, marker_color = 'black', name = "Successful Invasion", xaxis = 'x5'), row = 1, col = 1)
fig.add_trace(go.Histogram(x = sum_init_A_no_inv3, nbinsx = 100, marker_color = 'rgb(150, 150, 150)', name = "Unuccessful Invasion", opacity=0.60, xaxis = 'x5'), row = 1, col = 1)
fig.add_trace(go.Histogram(x = sum_init_A_inv4, nbinsx = 100, marker_color = 'black', showlegend = False, xaxis = 'x5'), row = 2, col = 1)
fig.add_trace(go.Histogram(x = sum_init_A_no_inv4, nbinsx = 50, marker_color = 'rgb(150, 150, 150)', showlegend = False, opacity=0.60, xaxis = 'x5'), row = 2, col = 1)
fig.add_trace(go.Histogram(x = sum_init_A_inv8, nbinsx = 70, marker_color = 'black', showlegend = False, xaxis = 'x5'), row = 3, col = 1)
fig.add_trace(go.Histogram(x = sum_init_A_no_inv8, nbinsx = 150, marker_color = 'rgb(150, 150, 150)', showlegend = False, opacity=0.60), row = 3, col = 1)

# update axis data
fig.update_xaxes(showline=True, linecolor='black')
fig.update_yaxes(showline=True, linecolor='black')
fig.update_layout(barmode='overlay', plot_bgcolor = 'rgba(0,0,0,0)', 
                  font = dict(size = 20, color = 'black'),
                  width = 1000,
                  height = 700,
                  yaxis2 = dict(title = "Invasion Frequency"),
                  xaxis3 = dict(title = "Total Initial Connected Pollinator Abundance"),
                  legend=dict(
                     yanchor = 'bottom',
                     y=0.78,
                     xanchor = 'center',
                     x=0.75))

fig.add_annotation(xref="x domain",yref="y domain",x=0.5, y=1.2, showarrow=False,
                   text="Spec High R", row=1, col=1)
fig.add_annotation(xref="x domain",yref="y domain",x=0.5, y=1.2, showarrow=False,
                   text="Spec High R&P", row=2, col=1)
fig.add_annotation(xref="x domain",yref="y domain",x=0.5, y=1.2, showarrow=False,
                   text="Gen High R&P", row=3, col=1)


fig.show()
fig.write_image('fig1bcd.png', width = 800, height = 650, scale = 10)
```

```python
# boxplots of total initial pollinator abundance

fig = make_subplots(rows=1, cols=3,
                    y_title = ("Pollinator Abundance"),
                    subplot_titles = ("Spec High R", "Spec High P&R", "Gen High P&R"))

# add all boxplots
fig.add_trace(go.Box(y = sum_init_A_inv3, marker_color = 'black', showlegend = False, name = "Invasion Success"), row = 1, col = 1)
fig.add_trace(go.Box(y = sum_init_A_no_inv3, marker_color = 'rgb(150, 150, 150)', showlegend = False, name = "Invasion Failure"), row = 1, col = 1)
fig.add_trace(go.Box(y = sum_init_A_inv4, marker_color = 'black', showlegend = False, name = "Invasion Success"), row = 1, col = 2)
fig.add_trace(go.Box(y = sum_init_A_no_inv4, marker_color = 'rgb(150, 150, 150)', showlegend = False, name = "Invasion Failure"), row = 1, col = 2)
fig.add_trace(go.Box(y = sum_init_A_inv8, marker_color = 'black', showlegend = False, name = "Invasion Success"), row = 1, col = 3)
fig.add_trace(go.Box(y = sum_init_A_no_inv8, marker_color = 'rgb(150, 150, 150)', showlegend = False, name = "Invasion Failure"), row = 1, col = 3)

# update axis data
fig.update_xaxes(showline=True, linecolor='black')
fig.update_yaxes(showline=True, linecolor='black')
fig.update_layout(barmode='overlay', plot_bgcolor = 'rgba(0,0,0,0)', 
                  font = dict(size = 15, color = 'black'), 
                  legend=dict(
                     yanchor = 'bottom',
                     y=0.78,
                     xanchor = 'center',
                     x=0.75))
fig.show()
```

```python
# boxplots for change in network structure

#only for networks successfully invaded that have about 40 species
network_data_m3_l1_inv = network_data_m3_l1[network_data_m3_l1.InvP_inv == 1]
network_data_m3_l1_inv_s40 = network_data_m3_l1_inv[network_data_m3_l1_inv.S < 50]

# add all boxplots
fig = make_subplots(rows=2, cols=2)
fig.add_trace(go.Box(y = network_data_m3_l1_inv_s40.init_avg_indirect_quantity, marker_color = 'black', showlegend = False, name = "Before"), row = 1, col = 1)
fig.add_trace(go.Box(y = network_data_m3_l1_inv_s40.fin_avg_indirect_quantity, marker_color = 'rgb(150, 150, 150)', showlegend = False, name = "After"), row = 1, col = 1)
fig.add_trace(go.Box(y = network_data_m3_l1_inv_s40.init_avg_indirect_quality, marker_color = 'black', showlegend = False, name = "Before"), row = 1, col = 2)
fig.add_trace(go.Box(y = network_data_m3_l1_inv_s40.fin_avg_indirect_quality, marker_color = 'rgb(150, 150, 150)', showlegend = False, name = "After"), row = 1, col = 2)
fig.add_trace(go.Box(y = network_data_m3_l1_inv_s40.init_wNODF, marker_color = 'black', showlegend = False, name = "Before"), row = 2, col = 1)
fig.add_trace(go.Box(y = network_data_m3_l1_inv_s40.fin_wNODF, marker_color = 'rgb(150, 150, 150)', showlegend = False, name = "After"), row = 2, col = 1)
fig.add_trace(go.Box(y = network_data_m3_l1_inv_s40.init_mod, marker_color = 'black', showlegend = False, name = "Before"), row = 2, col = 2)
fig.add_trace(go.Box(y = network_data_m3_l1_inv_s40.fin_mod, marker_color = 'rgb(150, 150, 150)', showlegend = False, name = "After"), row = 2, col = 2)


# update axis data
fig.update_xaxes(showline=True, linecolor='black')
fig.update_yaxes(showline=True, linecolor='black')
fig.update_layout(barmode='overlay', plot_bgcolor = 'rgba(0,0,0,0)', 
                  font = dict(size = 15, color = 'black'), 
                  legend=dict(
                     yanchor = 'bottom',
                     y=0.78,
                     xanchor = 'center',
                     x=0.75))
fig.show()
fig.write_image('fig3.png', width = 800, height = 650, scale = 10)
```

```python
# scatter plot for effect on native plant and pollinator abundance

data = {'P': network_data_m3_l1_i4['P'], 'P_diff' : network_data_m3_l1_i4['P_diff'], 'A_diff' : network_data_m3_l1_i4['A_diff'], 'P_pol_diff' : network_data_m3_l1_i4['P_pol_diff'], 'A_visit_diff' : network_data_m3_l1_i4['A_visit_diff']}

# effect on native abundance
fig = make_subplots(rows=2, cols=1, shared_xaxes=True, x_title = "Number of Plant Species", y_title = "Proportional Abundance Difference")
fig.add_trace(go.Scatter(x = data['P'], y = data['P_diff'], mode='markers', marker = dict(color = 'rgb(150, 150, 150)'), name = "Native Plants"), row = 1, col = 1)
fig.add_trace(go.Scatter(x = data['P'], y = data['A_diff'], mode='markers', marker = dict(color = 'black'), name = "Native Pollinators"), row = 2, col = 1)

# update axis data
fig.update_xaxes(showline=True, linecolor='black')
fig.update_yaxes(showline=True, linecolor='black')
fig.update_layout(plot_bgcolor = 'rgba(0,0,0,0)', 
                  font = dict(size = 16, color = 'black'),
                  legend=dict(
                    yanchor="top",
                      y = 0.5,
                    xanchor="right",
))
fig.show()
fig.write_image('fig4.png', width = 800, height = 650, scale = 10)
```
